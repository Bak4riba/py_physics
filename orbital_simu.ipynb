Programa para cálculo do da velocidade em x e y durante a interação de dois corpos. Desenvolvi o programa para fins de estudo de simulação em python.


```python
import numpy as np

print("Determinando as condições iniciais.............\n")
# Condições iniciais
G = 6.67430e-11  # Constante gravitacional (m^3 kg^-1 s^-2)
m1 = 1.989e30    # Massa do Sol (kg)
m2 = 5.972e24    # Massa da Terra (kg)

# Posições iniciais (metros)
x1, y1 = 0, 0      # Sol na origem
x2, y2 = 1.496e11, 0  # Terra a 1 UA no eixo x

r_x = x2 - x1
r_y = y2 - y1
r = np.sqrt(r_x**2 + r_y**2)
a1x = -G * m2 / r**3 * r_x
a1y = -G * m2 / r**3 * r_y

a2x = G * m1 / r**3 * r_x
a2y = G * m1 / r**3 * r_y

# Imprima os resultados
print(f"Distancia em X: {r_x}\n")
print(f"Distancia em Y: {r_y}\n")
print(f"Posição do corpo 2: ({x2:.3e}, {y2:.3e}) m \n")
print(f"Posição do corpo 1: ({x1:.3e}, {y1:.3e}) m \n")
print(f"Aceleração do corpo 1: ({a1x:.3e}, {a1y:.3e}) m/s^2 \n")
print(f"Aceleração do corpo 2: ({a2x:.3e}, {a2y:.3e}) m/s^2 \n")
```
