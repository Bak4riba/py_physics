## Analisando MRU
---
Vamos analisar um problema de dinâmica, que é o movimento uniforme, ou seja, o movimento de um corpo que varia de maneira conforme no espaço e nos mesmos intervalos de tempo.

### Problema 1
---
Um móvel desloca-se ao longo de uma trajetória retilínea com velocidade constante. No instante inicial $t_0=0s$, o móvel encontra-se na posição $x_0=12m$

A velocidade escalar do movimento é $v=3,5 m/s$ no sentido positivo da trajetória.

No codigo a seguir mostramos o gráfico que demonstra essa equação.

```python
import numpy as np

t = 0; # tempo inicial
x = 12; # x inicial
v = 3.5; # velocidade
print("Tempo(s) | Distancia (m) | Velocidade (m/s)");
while t <= 10:
  print(f"{t} | {x} |{v}")
  t = t + 1;
  x = x + v;

```
